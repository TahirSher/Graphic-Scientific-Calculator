```python
# Install SymPy for symbolic mathematics (optional, if not installed)
# !pip install sympy

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from math import sin, cos, tan, sqrt, log, exp, pi

# Simple scientific calculator functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Error! Division by zero."

# Advanced scientific functions
def sine(x):
    return sin(x)

def cosine(x):
    return cos(x)

def tangent(x):
    return tan(x)

def square_root(x):
    return sqrt(x)

def logarithm(x):
    return log(x)

def exponential(x):
    return exp(x)

# Plotting function
def plot_function(func, x_min=-10, x_max=10):
    x = np.linspace(x_min, x_max, 400)
    y = func(x)
    plt.plot(x, y)
    plt.title(f'Plot of {func.__name__}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()

# Symbolic differentiation and integration using SymPy
def symbolic_operations():
    x = sp.symbols('x')
    expr = input("Enter an expression in terms of x (e.g., x**2 + 2*x + 1): ")
    expr = sp.sympify(expr)

    derivative = sp.diff(expr, x)
    integral = sp.integrate(expr, x)

    print(f"Derivative of {expr}: {derivative}")
    print(f"Indefinite Integral of {expr}: {integral}")

# Main menu
def calculator():
    print("Select operation:")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Sine")
    print("6. Cosine")
    print("7. Tangent")
    print("8. Square Root")
    print("9. Logarithm (natural log)")
    print("10. Exponential (e^x)")
    print("11. Plot function")
    print("12. Symbolic Differentiation and Integration")

    choice = input("Enter choice: ")

    if choice in ['1', '2', '3', '4']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if choice == '1':
            print(f"{num1} + {num2} = {add(num1, num2)}")
        elif choice == '2':
            print(f"{num1} - {num2} = {subtract(num1, num2)}")
        elif choice == '3':
            print(f"{num1} * {num2} = {multiply(num1, num2)}")
        elif choice == '4':
            print(f"{num1} / {num2} = {divide(num1, num2)}")

    elif choice in ['5', '6', '7', '8', '9', '10']:
        num = float(input("Enter number: "))

        if choice == '5':
            print(f"sin({num}) = {sine(num)}")
        elif choice == '6':
            print(f"cos({num}) = {cosine(num)}")
        elif choice == '7':
            print(f"tan({num}) = {tangent(num)}")
        elif choice == '8':
            print(f"sqrt({num}) = {square_root(num)}")
        elif choice == '9':
            print(f"log({num}) = {logarithm(num)}")
        elif choice == '10':
            print(f"exp({num}) = {exponential(num)}")

    elif choice == '11':
        print("Select a function to plot:")
        print("1. Sine")
        print("2. Cosine")
        print("3. Tangent")
        print("4. Exponential")
        print("5. Square Root")
        func_choice = input("Enter function choice: ")

        if func_choice == '1':
            plot_function(np.sin)
        elif func_choice == '2':
            plot_function(np.cos)
        elif func_choice == '3':
            plot_function(np.tan)
        elif func_choice == '4':
            plot_function(np.exp)
        elif func_choice == '5':
            plot_function(np.sqrt)

    elif choice == '12':
        symbolic_operations()

    else:
        print("Invalid input")

# Run the calculator
calculator()
```

```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from math import sin, cos, tan, sqrt, log, exp, pi

# Title of the app
st.title('Scientific Calculator with Plotting')

# Calculator functionality
def calculator():
    st.subheader("Basic Operations")

    # Input numbers and operation
    num1 = st.number_input("Enter first number", step=1e-6, format="%.6f")
    num2 = st.number_input("Enter second number", step=1e-6, format="%.6f")
    operation = st.selectbox("Choose an operation", ("Add", "Subtract", "Multiply", "Divide"))

    if operation == "Add":
        result = num1 + num2
        st.write(f"Result: {num1} + {num2} = {result}")
    elif operation == "Subtract":
        result = num1 - num2
        st.write(f"Result: {num1} - {num2} = {result}")
    elif operation == "Multiply":
        result = num1 * num2
        st.write(f"Result: {num1} * {num2} = {result}")
    elif operation == "Divide":
        if num2 != 0:
            result = num1 / num2
            st.write(f"Result: {num1} / {num2} = {result}")
        else:
            st.error("Error! Division by zero.")

def scientific_functions():
    st.subheader("Scientific Functions")

    # Input number
    num = st.number_input("Enter a number for scientific calculation", step=1e-6, format="%.6f")
    func = st.selectbox("Choose a function", ("Sine", "Cosine", "Tangent", "Square Root", "Logarithm", "Exponential"))

    if func == "Sine":
        result = sin(num)
        st.write(f"sin({num}) = {result}")
    elif func == "Cosine":
        result = cos(num)
        st.write(f"cos({num}) = {result}")
    elif func == "Tangent":
        result = tan(num)
        st.write(f"tan({num}) = {result}")
    elif func == "Square Root":
        if num >= 0:
            result = sqrt(num)
            st.write(f"sqrt({num}) = {result}")
        else:
            st.error("Error! Cannot compute square root of a negative number.")
    elif func == "Logarithm":
        if num > 0:
            result = log(num)
            st.write(f"log({num}) = {result}")
        else:
            st.error("Error! Logarithm only defined for positive numbers.")
    elif func == "Exponential":
        result = exp(num)
        st.write(f"exp({num}) = {result}")

def plot_function():
    st.subheader("Plot a Function")

    # Choose a function to plot
    func_choice = st.selectbox("Select a function to plot", ("Sine", "Cosine", "Tangent", "Exponential", "Square Root"))
    x_min = st.number_input("Enter the minimum x value", -100, 0, -10)
    x_max = st.number_input("Enter the maximum x value", 0, 100, 10)

    x = np.linspace(x_min, x_max, 400)

    if func_choice == "Sine":
        y = np.sin(x)
        st.write("Plot of sin(x)")
    elif func_choice == "Cosine":
        y = np.cos(x)
        st.write("Plot of cos(x)")
    elif func_choice == "Tangent":
        y = np.tan(x)
        st.write("Plot of tan(x)")
    elif func_choice == "Exponential":
        y = np.exp(x)
        st.write("Plot of exp(x)")
    elif func_choice == "Square Root":
        y = np.sqrt(np.clip(x, 0, None))  # Clip to avoid negative values for sqrt
        st.write("Plot of sqrt(x)")

    # Plot the graph
    fig, ax = plt.subplots()
    ax.plot(x, y)
    ax.grid(True)
    st.pyplot(fig)

def symbolic_operations():
    st.subheader("Symbolic Differentiation and Integration")

    x = sp.symbols('x')
    expr = st.text_input("Enter an expression in terms of x (e.g., x**2 + 2*x + 1)")
    if expr:
        expr = sp.sympify(expr)
        derivative = sp.diff(expr, x)
        integral = sp.integrate(expr, x)

        st.write(f"Expression: {expr}")
        st.write(f"Derivative of the expression: {derivative}")
        st.write(f"Indefinite integral of the expression: {integral}")

# Layout of the calculator
st.sidebar.title("Choose a feature")
feature = st.sidebar.radio("Select a feature", ("Basic Calculator", "Scientific Functions", "Plot Functions", "Symbolic Calculations"))

if feature == "Basic Calculator":
    calculator()
elif feature == "Scientific Functions":
    scientific_functions()
elif feature == "Plot Functions":
    plot_function()
elif feature == "Symbolic Calculations":
    symbolic_operations()
```
